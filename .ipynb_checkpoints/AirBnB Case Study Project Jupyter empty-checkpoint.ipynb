```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airbnb = pd.read_csv('./AB_NYC_2019.csv')
airbnb.head(10)
```

# Question #2: How many neighborhood groups are available and which shows up the most?
48,895 
Manhattan

```python
airbnb.keys()
airbnb.count()['neighbourhood_group']
airbnb.groupby(['neighbourhood_group']).neighbourhood_group.count().sort_values(ascending=False)
```

# Question #3: Are private rooms the most popular in manhattan?
No

```python
airbnb.query('neighbourhood_group == "Manhattan"').room_type.value_counts()
```

# Question #4: Which hosts are the busiest and based on their reviews?
Michael, but David is creeping up there

```python
airbnb.groupby('host_name')[['number_of_reviews', 'reviews_per_month']].sum().nlargest(5, 'number_of_reviews')
```

# Question #5: Which neighorhood group has the highest average price?
Manhattan has the highest average price.

```python
airbnb.groupby('neighbourhood_group')[[ 'price']].describe()
```

# Question #6: Which neighborhood group has the highest total price?
Brooklyn, Manhattan, Queens

```python
airbnb.groupby('neighbourhood_group')[[ 'price']].describe()
```

# Question #7: Which top 5 hosts have the highest total price?
Sonder(NYC)

```python
airbnb.groupby('host_name')[['price']].sum().nlargest(5, 'price')
```

# Question #8: Who currently has no (zero) availability with a review count of 100 or more?

```python
t=airbnb.query('availability_365 == 0 and number_of_reviews >= 100')[['host_name', 'number_of_reviews', 'availability_365']]
print(t)
```

# Question #9: What host has the highest total of prices and where are they located?
Sonder and they live in the Financial District

```python
airbnb.groupby(['host_name', 'neighbourhood'])[[ 'price']].sum('price').nlargest(1, 'price')
```

# Question #10: When did Danielle from Queens last receive a review?
07/08/2019

```python
airbnb.query('host_name == "Danielle" and neighbourhood_group == "Queens"').sort_values(by=['last_review'], ascending=False)[['host_name', 'last_review']]
```

## Further Questions
